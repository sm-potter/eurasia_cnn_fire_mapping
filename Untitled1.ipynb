The image generator gives different IOU, for one fire plot the non chunked dnbr, then re-work the chunked dnbr back to original and compare

```python
import os
import numpy as np
import rioxarray
import matplotlib.pyplot as plt
import random

# Input path where the original TIFF files are stored
in_path = '/explore/nobackup/people/spotter5/cnn_mapping/nbac_training/nbac_monthly_ndsi_sliding'
# Path where the 128x128 chunks are stored (output path from previous script)
chunk_path = '/explore/nobackup/people/spotter5/cnn_mapping/nbac_training/nbac_monthly_ndsi_sliding_subs_0_128'
size = 128  # Chunk size (128x128)

# Function to load the original TIFF file and extract the dNBR and target (band 7)
def load_original_tif(file_name):
    tif_path = os.path.join(in_path, file_name)
    img = rioxarray.open_rasterio(tif_path).to_numpy()
    img = np.moveaxis(img, 0, 2)  # Move bands to last axis
    dnbr = img[:, :, 0]  # First band (dNBR)
    target = img[:, :, -1]  # Last band (target)
    return dnbr, target, img.shape

# Function to load and reassemble 128x128 chunks into a full image using numpy
def reassemble_image(file_name, original_shape):
    # Extract the TIFF ID from file_name, e.g., for 'median_1.tif' the ID is 1
    tif_id = file_name.split('_')[1].replace('.tif', '')

    # Create empty arrays for reassembling the dNBR and target (full image size)
    reassembled_dnbr = np.zeros((original_shape[0], original_shape[1]))
    reassembled_target = np.zeros((original_shape[0], original_shape[1]))

    # Loop through the directory of chunks and select the ones associated with the current file_name
    chunk_files = [f for f in os.listdir(chunk_path) if f.endswith(f'_{tif_id}.npy')]  # Dynamically select files based on file_name

    for chunk_file in chunk_files:
        # Extract the chunk's row (i) and column (j) position from the filename
        i, j = map(int, chunk_file.split('_')[:2])
        
        # Load the chunk and extract the dNBR (1st band) and target (last band)
        chunk = np.load(os.path.join(chunk_path, chunk_file))
        dnbr_chunk = chunk[:, :, 0]  # First band (dNBR)
        target_chunk = chunk[:, :, -1]  # Last band (target)
        
        # Calculate the starting and ending indices for placement in the reassembled image
        row_start = i * size
        row_end = row_start + size
        col_start = j * size
        col_end = col_start + size
        
        # Place the chunks in the appropriate location of the reassembled image
        reassembled_dnbr[row_start:row_end, col_start:col_end] = dnbr_chunk
        reassembled_target[row_start:row_end, col_start:col_end] = target_chunk
    
    return reassembled_dnbr, reassembled_target

# Plot function to compare original and reassembled images
def plot_comparison(original_dnbr, original_target, reassembled_dnbr, reassembled_target):
    fig, ax = plt.subplots(2, 2, figsize=(12, 10))

    ax[0, 0].imshow(original_dnbr, cmap='coolwarm')
    ax[0, 0].set_title('Original dNBR')
    
    ax[0, 1].imshow(original_target, cmap='gray')
    ax[0, 1].set_title('Original Target')

    ax[1, 0].imshow(reassembled_dnbr, cmap='coolwarm')
    ax[1, 0].set_title('Reassembled dNBR')
    
    ax[1, 1].imshow(reassembled_target, cmap='gray')
    ax[1, 1].set_title('Reassembled Target')

    plt.tight_layout()
    plt.show()

# Example file to process
file_name = 'median_1.tif'  # Replace with your example file, e.g., 'median_2.tif'

# Function to randomly select a .tif file from the directory
# def select_random_tif_file():
#     tif_files = [f for f in os.listdir(in_path) if f.endswith('.tif')]
#     if not tif_files:
#         raise ValueError("No .tif files found in the directory.")
#     return random.choice(tif_files)

# Select a random .tif file
# file_name = select_random_tif_file()

# Load the original image
original_dnbr, original_target, original_shape = load_original_tif(file_name)

# Reassemble the image from 128x128 chunks
reassembled_dnbr, reassembled_target = reassemble_image(file_name, original_shape)

# Plot comparison of original vs reassembled images
plot_comparison(original_dnbr, original_target, reassembled_dnbr, reassembled_target)
```
